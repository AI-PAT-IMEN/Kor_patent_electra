```python
def make(params_2):
  
    lst_a = []
    lst_b = []
    lst_c = []

    params_2 = params_2.split('\n')
    for idx,i in enumerate(params_2):    
        if idx == 0:
            lst_c.append('')

        if ':' in i:
            lst_a.append(i)

            if idx == 0:
                lst_b.append('')
                continue

            if (':' in lst_c[-1]) & (':' in i):
                lst_b.append('')

        else:

            lst_b.append(i)

        lst_c.append(i)

    params = {}
    for i,j in zip(lst_a,lst_b):
        params[i.split(':')[0].strip()] = j
        
    return params
        
```
